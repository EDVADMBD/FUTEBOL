```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados fictícios de jogadores, seleções e gols marcados na Copa América 2024
dados_jogadores = {
    "Jogador": ["Lionel Messi (Argentina)", "Vinícius Jr. (Brasil)", "Luis Suárez (Uruguai)", "Alexis Sánchez (Chile)", "Darwin Núñez (Uruguai)", "Lautaro Martínez (Argentina)", "Enner Valencia (Equador)"],
    "Gols Marcados": [6, 5, 4, 3, 4, 6, 4]  # Exemplos fictícios de gols marcados
}

# Criar o DataFrame
df_jogadores = pd.DataFrame(dados_jogadores)

# Ordenar os jogadores pelo número de gols marcados
df_jogadores = df_jogadores.sort_values(by='Gols Marcados', ascending=False)

# Criar o gráfico de barras horizontais
plt.figure(figsize=(10, 6))
plt.barh(df_jogadores['Jogador'], df_jogadores['Gols Marcados'], color='lightblue')
plt.xlabel('Número de Gols Marcados')
plt.ylabel('Jogador')
plt.title('Potenciais Melhores Atacantes na Copa América 2024')
plt.gca().invert_yaxis()  # Inverter o eixo y para que o jogador com mais gols fique no topo
plt.tight_layout()

# Mostrar o gráfico
plt.show()



```

```python
#seleção historicamente mais provável de vencer a Copa América 2024
import pandas as pd
import matplotlib.pyplot as plt

# Dados das campanhas das 5 melhores seleções em Copa América
campanhas_data = {
    "País": ["Argentina", "Brasil", "Uruguai", "Chile", "Colômbia"],
    "Títulos": [15, 9, 15, 2, 1],
    "Vice-campeonatos": [14, 12, 6, 4, 1],
    "Terceiros lugares": [5, 7, 9, 5, 5],
    "Quartos lugares": [4, 3, 6, 5, 2],
    "Total Participações": [42, 37, 45, 39, 24]
}

# Criar o DataFrame
df_campanhas_copa_america = pd.DataFrame(campanhas_data)

# Calcular a probabilidade de vitória com base no histórico de títulos
df_campanhas_copa_america["Probabilidade de Vitória (%)"] = (df_campanhas_copa_america["Títulos"] / df_campanhas_copa_america["Total Participações"]) * 100

# Exibir o DataFrame atualizado
print(df_campanhas_copa_america)

# Plotar o gráfico de probabilidades
plt.figure(figsize=(10, 6))
plt.bar(df_campanhas_copa_america["País"], df_campanhas_copa_america["Probabilidade de Vitória (%)"], color='lightblue')
plt.xlabel('País')
plt.ylabel('Probabilidade de Vitória (%)')
plt.title('Projeção de Vitória na Copa América 2024')
plt.show()

# Seleção com maior probabilidade de vitória
selecao_mais_probabilidade = df_campanhas_copa_america.loc[df_campanhas_copa_america["Probabilidade de Vitória (%)"].idxmax()]

print(f"Seleção com maior probabilidade de vitória: {selecao_mais_probabilidade['País']}")
print(f"Probabilidade de Vitória: {selecao_mais_probabilidade['Probabilidade de Vitória (%)']:.2f}%")
```
