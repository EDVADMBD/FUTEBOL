```python
#Grafico das Melhores Seleções da Eurocopa
import pandas as pd
import matplotlib.pyplot as plt

# Dados fictícios das melhores seleções da Eurocopa
selecoes = ["Bélgica", "França", "Inglaterra", "Portugal", "Espanha",
            "Itália", "Alemanha", "Países Baixos (Holanda)", "Dinamarca", "Croácia"]
rankings_ficticios = [1, 2, 4, 5, 6, 7, 8, 9, 10, 11]  # Exemplos fictícios de ranking

# Criar DataFrame com os dados
df_rankings = pd.DataFrame({
    "Seleção": selecoes,
    "Ranking FIFA": rankings_ficticios
})

# Ordenar o DataFrame pelo ranking FIFA (do melhor para o pior)
df_rankings = df_rankings.sort_values(by="Ranking FIFA", ascending=True)

# Plotar o gráfico de barras
plt.figure(figsize=(10, 6))
plt.barh(df_rankings["Seleção"], df_rankings["Ranking FIFA"], color='skyblue')
plt.xlabel('Ranking FIFA')
plt.ylabel('Seleção')
plt.title('Melhores Seleções da Eurocopa (Fictício)')
plt.gca().invert_yaxis()  # Inverter o eixo y para que a seleção melhor classificada fique no topo
plt.tight_layout()

# Mostrar o gráfico
plt.show()
```

```python
#seleções com melhor ataque e melhor defesa com base nos dados fornecidos da Eurocopa
import pandas as pd

# Dados fictícios de gols marcados e sofridos pelas seleções na Eurocopa
dados_eurocopa = {
    "Seleção": ["Alemanha", "Espanha", "França", "Itália", "Países Baixos (Holanda)"],
    "Gols Marcados": [35, 30, 28, 25, 20],  # Exemplos fictícios de gols marcados
    "Gols Sofridos": [15, 12, 10, 18, 20]   # Exemplos fictícios de gols sofridos
}

# Criar o DataFrame
df_gols = pd.DataFrame(dados_eurocopa)

# Calcular o melhor ataque (maior número de gols marcados)
melhor_ataque = df_gols[df_gols['Gols Marcados'] == df_gols['Gols Marcados'].max()]['Seleção'].values[0]

# Calcular a melhor defesa (menor número de gols sofridos)
melhor_defesa = df_gols[df_gols['Gols Sofridos'] == df_gols['Gols Sofridos'].min()]['Seleção'].values[0]

# Mostrar resultados
print(f"Melhor ataque na Eurocopa: {melhor_ataque} com {df_gols['Gols Marcados'].max()} gols marcados.")
print(f"Melhor defesa na Eurocopa: {melhor_defesa} com {df_gols['Gols Sofridos'].min()} gols sofridos.")
```

```python
#criar um gráfico em porcentagem que visualize a participação das seleções com melhor e pior ataque
import matplotlib.pyplot as plt

# Dados fictícios de gols marcados pelas seleções na Eurocopa
selecoes = ["Alemanha", "Espanha", "França", "Itália", "Países Baixos (Holanda)"]
gols_marcados = [35, 30, 28, 25, 20]  # Exemplos fictícios de gols marcados

# Calcular porcentagens de participação com base nos gols marcados
total_gols = sum(gols_marcados)
porcentagens = [(gols / total_gols) * 100 for gols in gols_marcados]

# Criar o gráfico de pizza
plt.figure(figsize=(8, 8))
plt.pie(porcentagens, labels=selecoes, autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightgreen', 'lightskyblue', 'lightpink'], shadow=True, startangle=140)
plt.title('Participação das Seleções com Melhor Ataque na Eurocopa')
plt.axis('equal')  # Garantir que o gráfico de pizza seja desenhado como um círculo
plt.tight_layout()

# Mostrar o gráfico
plt.show()
```
